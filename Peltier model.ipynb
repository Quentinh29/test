```python
# Evaluate cooling powers from peltier
while Tp[i-1] > Tmelt:    
    Tmould = np.append(Tmould,[Twaste + Pout[i-1]*R_mould])
    Ppelt2 = np.arange(0,70.1,0.1)
    TwarmPeltier = np.append(TwarmPeltier, [Twaste + Pout[i-1]*R_warm])
    dT = np.append(dT, [TwarmPeltier[i]-Tpeltier[i-1]])
    dTeval = (np.zeros(len(Ppelt2))) + dT[i]
    c = Peltier_A
    z_A = poly2Dreco(dTeval,Ppelt2,c)                                          # z_A = feval(peltier_A,[dTeval(:),Ppelt2(:)])
    Ppelt2 = np.delete(Ppelt2, np.argwhere(z_A>factorA*max(Z_A)))
    Ppelt = np.append(Ppelt, [max(Ppelt2)])         #[max(Ppelt2)]             # select highest cooling powers possible and index
    c = Peltier_Pout
    Pout = np.append(Pout, [poly2Dreco(dT[i], Ppelt[i], c)])
    Prad = episolon_alu*A_insert*Boltz*(Tmould[i]**4-Tp[i-1]**4)
    Pconv = hconv*A_insert*(Tmould[i]-Tp[i-1])
    Qcool = (Ppelt[i] - Prad - Pconv) * dt
    Tpeltier = np.append(Tpeltier, [Tp[i-1] - (Qcool/(c_alu*m_insert+c_water*m_water+c_solids*m_solids))-(Ppelt[i]*R_cold)])
    Tp = np.append(Tp, [Tpeltier[i]+(Ppelt[i]*R_cold)])
    t = np.append(t, [t[i-1]+dt])    
    i = i+1 
```

```python
# Ice freezing phase
Tp[i-1]=273.15                                                                 # Tp(i-1)= 273.15
Tp = np.append(Tp, 273.15)                                                     # Tp(i)= 273.15;
t = np.append(t, t[i-1]+(m_water*dH_melt)/Qcool)                               # t(i) = t(i-1)+(m_water*dH_melt)/Qcool;
Pout = np.append(Pout, Pout[i-1])                                              # Pout(i)=Pout(i-1);
Tmould = np.append(Tmould, Tmould[i-1])                                        # Tmould(i)=Tmould(i-1);
dT[i-1] = Tmould[i-1] - Tp[i]                                                  # dT(i-1)= Tmould(i-1)-Tp(i);
dT = np.append(dT, Tmould[i]-Tp[i])                                            # dT(i)= Tmould(i)-Tp(i);
Ppelt = np.append(Ppelt , Ppelt[i-1])                                          # Ppelt(i) = Ppelt(i-1);
Tpeltier = np.append(Tpeltier, Tp[i]-(Ppelt[i]*h_pad)/(l_pad2**2*hpad))        # Tpeltier(i) = Tp(i)-(Ppelt(i)*h_pad)/(l_pad2^2*hpad);
TwarmPeltier = np.append(TwarmPeltier, TwarmPeltier[i-1])                      # TwarmPeltier(i)=TwarmPeltier(i-1);
Tpeltier[i-1] = Tpeltier[i]                                                    # Tpeltier(i-1)=Tpeltier(i);
i = i +1
i_freeze = i
```

```python
# Ice Cooling phase
while i-2<i_freeze or Tp[i-2]-Tp[i-1] > 0.01:                                  # while (i-2<i_freeze || Tp(i-2)-Tp(i-1)>0.01)
    # evaluate cooling powers from peltier
    Tmould = np.append(Tmould, Twaste + Pout[i-1]*R_mould)                     #((h_pad/hpad)+(h_mould/halu)*)/A_mould)
    Ppelt2 = np.arange(0,70.1,0.1)
    TwarmPeltier = np.append(TwarmPeltier, Twaste + Pout[i-1]*R_warm)
    dT = np.append(dT, [TwarmPeltier[i] - Tpeltier[i-1]])
    dTeval = (np.zeros(len(Ppelt2))) + dT[i]
    c = Peltier_A
    z_A = poly2Dreco(dTeval,Ppelt2,c)                                          # get Current needed for these cooling rates
    Ppelt2 = np.delete(Ppelt2, np.argwhere(z_A>factorA*max(Z_A)))              # delete all cooling powers above 85% max current peltier
    Ppelt = np.append(Ppelt, [max(Ppelt2)])                                    # select highest cooling powers possible and index
    c = Peltier_Pout
    Pout = np.append(Pout, [poly2Dreco(dT[i], Ppelt[i], c)])                   # Evaluate highest cooling rate at dT for waste energe
    Prad = episolon_alu*A_insert*Boltz*(Tmould[i]**4-Tp[i-1]**4)               # Radiation to insert
    Pconv = hconv*A_insert*(Tmould[i]-Tp[i-1])                                 # Convection to insert
    Qcool = (Ppelt[i] - Prad - Pconv) * dt                                     # Cooling heat left
    Tpeltier = np.append(Tpeltier, [Tp[i-1] - (Qcool/(c_alu*m_insert+c_water*m_water+c_solids*m_solids))-(Ppelt[i]*R_cold)]) # Resultant temp drop
    Tp = np.append(Tp, [Tpeltier[i]+(Ppelt[i]*R_cold)])
    t = np.append(t, [t[i-1]+dt])    
    i = i+1 
```
